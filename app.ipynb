```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('heart_disease.csv')
```

```python
data
```

```python
data.drop(columns=['education'],inplace=True)
```

```python
data
```

```python
data.shape
```

```python
data.dropna(inplace=True)
```

```python
data
```

```python
data['Gender'] = data['Gender'].replace(['Male','Female'],[1,0])
data['prevalentStroke'] = data['prevalentStroke'].replace(['yes','no'],[1,0])
```

```python
data.prevalentStroke.unique()
```

```python
data
```

```python
data['Heart_ stroke'] = data['Heart_ stroke'].replace(['yes','No'],[1,0])
```

```python
data['Heart_stroke'] = data['Heart_ stroke']
```

```python
data.Heart_stroke.unique()
```

```python
data
```

```python
data.drop(columns=['Heart_ stroke'],inplace=True)
```

```python
data
```

```python
data_input = data.drop(columns=['Heart_stroke'])
data_output = data['Heart_stroke']
```

```python
data_output.value_counts()
```

```python
from imblearn.over_sampling import SMOTE
```

```python
sm = SMOTE(random_state=2)
```

```python
data_input,data_output = sm.fit_resample(data_input,data_output.ravel())
```

```python
np.unique(data_output, return_counts=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
x_train,x_test,y_train,y_test=train_test_split(data_input,data_output,test_size=0.2)
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
input = np.array([0,59,0,0,0,0,1,0,258,138.5,85,34.55,65,103]).reshape(1, -1)
```

```python
x=model.predict(input)
```
